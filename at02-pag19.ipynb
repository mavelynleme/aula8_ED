```python
import networkx as nx
import matplotlib.pyplot as plt

# =====================================================
# ATIVIDADE 2 - GRAU DE ENTRADA E SAÍDA
# =====================================================

G = nx.DiGraph()

arestas = [
    (1, 2),
    (1, 3),
    (3, 2),
    (3, 4),
    (4, 3)
]

G.add_edges_from(arestas)

print("Grau de saída:")
for vertice in sorted(G.nodes()):
    print(f"d+({vertice}) = {G.out_degree(vertice)}")

print("\nGrau de entrada:")
for vertice in sorted(G.nodes()):
    print(f"d-({vertice}) = {G.in_degree(vertice)}")

pos = {
    1: (0, 0),
    2: (0, 2),
    3: (2, 1),
    4: (4, 2)
}

plt.figure(figsize=(7, 4))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color='lightblue',
    node_size=1200,
    font_weight='bold',
    font_color='black',
    edge_color='gray',
    arrows=True,
    arrowsize=25,
    width=2
)
plt.title("Atividade 2 - Grau de Entrada e Saída")
plt.show()
```
