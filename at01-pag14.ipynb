```python
import networkx as nx
import matplotlib.pyplot as plt

# =====================================================
# ATIVIDADE 1 - GRAU DOS VÉRTICES
# =====================================================

G = nx.Graph()

arestas = [
    (1, 2),
    (1, 5),
    (2, 3),
    (2, 5),
    (3, 4),
    (4, 5),
    (4, 6)
]

G.add_edges_from(arestas)

graus = dict(G.degree())

print("Graus de cada vértice:")
for vertice, grau in sorted(graus.items()):
    print(f"Vértice {vertice}: Grau {grau}")

maior_grau = max(graus.values())
menor_grau = min(graus.values())

vertices_maior = [v for v, g in graus.items() if g == maior_grau]
vertices_menor = [v for v, g in graus.items() if g == menor_grau]

print(f"\nMaior Grau: {maior_grau} (Pertence aos vértices: {vertices_maior})")
print(f"Menor Grau: {menor_grau} (Pertence ao(s) vértice(s): {vertices_menor})")

pos = {
    6: (0, 2),
    4: (1, 1.3),
    5: (2.3, 1.4),
    1: (3.4, 1.0),
    3: (1.3, 0),
    2: (2.5, 0.2)
}

plt.figure(figsize=(7, 4))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color='lightpink',
    node_size=1200,
    font_weight='bold',
    font_color='black',
    edge_color='gray',
    width=2
)
plt.title("Atividade 1 - Graus dos Vértices")
plt.show()
```
