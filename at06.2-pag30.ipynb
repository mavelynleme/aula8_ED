```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# =====================================================
# ATIVIDADE 6.2 - PÁGINA 30 - GRAFO DA DIREITA
# =====================================================

G = nx.Graph()

arestas = [
    (6, 4),
    (4, 5),
    (4, 3),
    (3, 2),
    (5, 2),
    (5, 1),
    (2, 1)
]

G.add_edges_from(arestas)

print("Lista de adjacência:")
for vertice in sorted(G.nodes()):
    vizinhos = sorted(list(G.neighbors(vertice)))
    print(f"{vertice}: {vizinhos}")

print("\nMatriz de adjacência:")
matriz = nx.to_pandas_adjacency(G, nodelist=sorted(G.nodes()), dtype=int)
display(matriz)

pos = {
    6: (0, 2),
    4: (1, 1.2),
    5: (2.2, 1.2),
    1: (3.2, 1),
    3: (1, 0),
    2: (2.2, 0)
}

plt.figure(figsize=(7, 4))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color='lightyellow',
    node_size=1200,
    font_weight='bold',
    edge_color='gray',
    width=2
)
plt.title("Atividade 6.2 - Página 30 - Grafo da Direita")
plt.show()
```
