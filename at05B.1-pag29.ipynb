```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# =====================================================
# ATIVIDADE 5B.1 - MATRIZ DO GRAFO SUPERIOR
# =====================================================

G = nx.Graph()

arestas = [
    (1, 2),
    (2, 3),
    (3, 4),
    (1, 5),
    (2, 6),
    (3, 7),
    (4, 8),
    (5, 6),
    (6, 7),
    (7, 8),
    (1, 4),
    (5, 8)
]

G.add_edges_from(arestas)

print("Lista de adjacência:")
for vertice in sorted(G.nodes()):
    vizinhos = sorted(list(G.neighbors(vertice)))
    print(f"{vertice}: {vizinhos}")

print("\nMatriz de adjacência:")
matriz = nx.to_pandas_adjacency(G, nodelist=sorted(G.nodes()), dtype=int)
display(matriz)

pos = {
    1: (0, 2),
    2: (1.5, 2),
    3: (3, 2),
    4: (4.5, 2),
    5: (0, 0),
    6: (1.5, 0),
    7: (3, 0),
    8: (4.5, 0)
}

plt.figure(figsize=(8, 4))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color='lightblue',
    node_size=1200,
    font_weight='bold',
    edge_color='gray',
    width=2
)
plt.title("Atividade 5B.1 - Grafo Superior")
plt.show()
```
