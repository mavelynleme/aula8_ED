```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# =====================================================
# ATIVIDADE 5A - GRAFO A PARTIR DA MATRIZ
# =====================================================

G = nx.Graph()

arestas = [
    (1, 2),
    (1, 3),
    (2, 3),
    (2, 5),
    (3, 4),
    (4, 5)
]

G.add_edges_from(arestas)

print("Arestas do grafo correspondente:")
for aresta in sorted(G.edges()):
    print(aresta)

print("\nLista de adjacência:")
for vertice in sorted(G.nodes()):
    vizinhos = sorted(list(G.neighbors(vertice)))
    print(f"{vertice}: {vizinhos}")

print("\nMatriz de adjacência:")
matriz = nx.to_pandas_adjacency(G, nodelist=sorted(G.nodes()), dtype=int)
display(matriz)

pos = {
    1: (0, 2),
    2: (1.5, 2),
    3: (1.5, 1),
    4: (3, 1),
    5: (3, 2)
}

plt.figure(figsize=(7, 4))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color='lightpink',
    node_size=1200,
    font_weight='bold',
    edge_color='gray',
    width=2
)
plt.title("Atividade 5A - Grafo Correspondente à Matriz")
plt.show()
```
