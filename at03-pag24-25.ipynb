```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# =====================================================
# ATIVIDADE 3 - LISTA E MATRIZ DE ADJACÊNCIA
# =====================================================

G = nx.Graph()

arestas = [
    ("v0", "v1"),
    ("v1", "v2"),
    ("v1", "v3"),
    ("v2", "v3"),
    ("v3", "v4")
]

G.add_edges_from(arestas)

print("Lista de adjacência:")
for vertice in sorted(G.nodes()):
    vizinhos = sorted(list(G.neighbors(vertice)))
    print(f"{vertice}: {vizinhos}")

print("\nMatriz de adjacência:")
matriz = nx.to_pandas_adjacency(G, nodelist=sorted(G.nodes()), dtype=int)
display(matriz)

pos = {
    "v0": (0, 2),
    "v1": (1, 1.2),
    "v2": (2, 0),
    "v3": (3, 1.2),
    "v4": (4, 2)
}

plt.figure(figsize=(7, 4))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color='lightgreen',
    node_size=1200,
    font_weight='bold',
    edge_color='gray',
    width=2
)
plt.title("Atividade 3 - Lista e Matriz de Adjacência")
plt.show()
```
