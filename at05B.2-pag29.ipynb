```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# =====================================================
# ATIVIDADE 5B.2 - MATRIZ DO GRAFO INFERIOR
# =====================================================

G = nx.Graph()

arestas = [
    (1, 3),
    (1, 5),
    (2, 4),
    (2, 6),
    (3, 5),
    (4, 6)
]

G.add_edges_from(arestas)

print("Lista de adjacência:")
for vertice in sorted(G.nodes()):
    vizinhos = sorted(list(G.neighbors(vertice)))
    print(f"{vertice}: {vizinhos}")

print("\nMatriz de adjacência:")
matriz = nx.to_pandas_adjacency(G, nodelist=sorted(G.nodes()), dtype=int)
display(matriz)

pos = {
    1: (0, 1.5),
    2: (1.5, 3),
    3: (3, 1.5),
    4: (3, 0),
    5: (1.5, -1.5),
    6: (0, 0)
}

plt.figure(figsize=(6, 5))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color='lightgreen',
    node_size=1200,
    font_weight='bold',
    edge_color='gray',
    width=2
)
plt.title("Atividade 5B.2 - Grafo Inferior")
plt.show()
```
