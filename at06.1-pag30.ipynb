```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# =====================================================
# ATIVIDADE 6.1 - PÁGINA 30 - GRAFO DA ESQUERDA
# =====================================================

G = nx.Graph()

arestas = [
    (0, 1),
    (0, 2),
    (0, 5),
    (1, 2),
    (2, 3),
    (2, 4),
    (3, 4),
    (3, 5)
]

G.add_edges_from(arestas)

print("Lista de adjacência:")
for vertice in sorted(G.nodes()):
    vizinhos = sorted(list(G.neighbors(vertice)))
    print(f"{vertice}: {vizinhos}")

print("\nMatriz de adjacência:")
matriz = nx.to_pandas_adjacency(G, nodelist=sorted(G.nodes()), dtype=int)
display(matriz)

pos = {
    0: (0, 2),
    1: (1, 1.2),
    2: (2, 2),
    5: (0, 0),
    3: (1.2, 0),
    4: (2, 0)
}

plt.figure(figsize=(7, 4))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color='lightpink',
    node_size=1200,
    font_weight='bold',
    edge_color='gray',
    width=2
)
plt.title("Atividade 6.1 - Página 30 - Grafo da Esquerda")
plt.show()
```
