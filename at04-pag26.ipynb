```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# =====================================================
# ATIVIDADE 4 - REDE DE RELACIONAMENTOS
# =====================================================

G = nx.Graph()

vertices = ["João", "Paulo", "Maria", "Joana", "Antonia", "Lili", "Raimundo"]

arestas = [
    ("João", "Paulo"),
    ("Paulo", "Maria"),
    ("Maria", "Joana"),
    ("Paulo", "Joana"),
    ("Joana", "Antonia"),
    ("Paulo", "Antonia"),
    ("Paulo", "Lili"),
    ("Lili", "Antonia")
]

G.add_nodes_from(vertices)
G.add_edges_from(arestas)

print("Lista de adjacência:")
for vertice in vertices:
    vizinhos = sorted(list(G.neighbors(vertice)))
    print(f"{vertice}: {vizinhos}")

print("\nMatriz de adjacência:")
matriz = nx.to_pandas_adjacency(G, nodelist=vertices, dtype=int)
display(matriz)

pos = {
    "João": (0, 1),
    "Paulo": (1.5, 1.5),
    "Maria": (2.2, 3),
    "Joana": (3.4, 2.4),
    "Antonia": (4.2, 1.3),
    "Lili": (2.8, 0),
    "Raimundo": (5.2, 0.8)
}

plt.figure(figsize=(8, 5))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color='lightyellow',
    node_size=1600,
    font_weight='bold',
    edge_color='gray',
    width=2
)
plt.title("Atividade 4 - Rede de Relacionamentos")
plt.show()
```
